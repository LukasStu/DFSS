```python
""" Bibliotheken importieren"""
import statistics
import numpy as np
from scipy.stats import norm
import scipy
from scipy.io import loadmat
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
%matplotlib inline  
```

```python
""" Einlesen der Messwerte"""
data = scipy.io.loadmat('VersuchsplanBlattfedern')['Data']

TetaOfen=data[:,0]
ZeitOfen=data[:,1]
TetaOel=data[:,2]
Elas=data[:,3]
```

a) Es handelt sich um einen vollfaktoriellen   $\{{\rm{2}}^{\rm{3}}}\$ Versuchsplan. Mit dem Versuchsplan können der konstanter Term, die linearen Terme und der Wechselwirkungsterm bestimmt werden. Da die Temperatur des Ölbads nicht gezielt beeinflusst werden kann, ergibt sich eine Funktion der Form 

$E = {b_0} + {b_1} \cdot {\vartheta _{OFEN}} + {b_2} \cdot {T_{OFEN}} + {b_3} \cdot {\vartheta _{OFEN}} \cdot {T_{OFEN}}$


b) Mithilfe der statsmodels-Bibliothek regstats kann eine Regressionsfunktion berechnet werden. 

```python
''' Statistische Bewertung '''
'''"""[0 0; 1 0; 0 1; 1 1] Modell anlegen""" '''
data_Dic={'X':TetaOfen, 'Y':ZeitOfen, 'Z':Elas}

model = ols("Z ~ X + Y + X*Y " , data_Dic).fit()
print(model.summary())
```

Es ergibt sich a0 = 52.1542, a1 = 0.0212, a2 = 1.6771 und a3 = - 0.0015. Die Güte der Regres-sionsfunktion ist nicht hoch. Das adjungierte Bestimmtheitsmaß beträgt 0.3011, ist also deut-lich kleiner als 1. Ein Normplot zeigt, dass die Residuen nicht normalverteilt sind. Außerdem deuten die hohen p-Values beim Signifikanztest darauf hin, dass das Modell nicht signifikante Terme aufweist. Das Modell wird trotzdem weiterverwendet.

```python
Residual= model.resid

#scipy.stats.probplot(Residual, rvalue=True, dist='norm', plot=plt)


plt.scatter(Residual, norm.cdf(Residual), marker='+')

[a,b] = np.polyfit(Residual, norm.cdf(Residual),  1)
R_Fit = np.polyval([a,b], Residual)

plt.plot(Residual, R_Fit, color= 'r')
plt.xlabel('Residuen r KN/mm2' )
plt.ylabel('Wahrscheinlichkeit P')
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.title('Wahrscheinlichkeitsnetz')
plt.ylim(bottom=-0.05)
```

```python
b1,b2,b3,b4=list(model._results.params)

'''% Regressionsfunktion als Grid berechnen'''
Teta,Zeit = np.meshgrid(np.arange(1100,1350,50), np.arange(2,12,2))
E = b1 + b2*Teta + b3*Zeit + b4*Teta*Zeit 


print ('ergibt sich folgende Darstellung:')

'''Grafische Darstellung'''
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(Teta,Zeit, E ,color='g', alpha=0.7)
ax.set_xlabel('Temperatur / °C') 
ax.set_ylabel('Verweilzeit / Min') 
ax.set_zlabel('Elastizitätsmodul / kN/mm²') 
```

c) Die Parameterkombinationen, mit denen der Zielwert erreicht werden kann, sind über die Regressionsgleichung gegeben.
$E = {a_0} + {a_1} \cdot {\vartheta _{OFEN}} + {a_2} \cdot {T_{OFEN}} + {a_3} \cdot {\vartheta _{OFEN}} \cdot {T_{OFEN}}$


Für eine gegebene Temperatur ${\vartheta _{OFEN}}$ kann die Verweilzeit  ${T_{OFEN}}$ über

$E - {a_0} - {a_1} \cdot {\vartheta _{OFEN}} = \left( {{a_2} + {a_3} \cdot {\vartheta _{OFEN}}} \right) \cdot {T_{OFEN}}$

berechnet werden zu

${T_{OFEN}} = \frac{{E - {a_0} - {a_1} \cdot {\vartheta _{OFEN}}}}{{{a_2} + {a_3} \cdot {\vartheta _{OFEN}}}} = \frac{{77 - 52.1542 - 0.0212 \cdot {\vartheta _{OFEN}}}}{{1.6771 - {\rm{ }}0.0015 \cdot {\vartheta _{OFEN}}}} = \frac{{{\rm{24}}{\rm{.8458}} - 0.0212 \cdot {\vartheta _{OFEN}}}}{{1.6771 - {\rm{ }}0.0015 \cdot {\vartheta _{OFEN}}}}$


Es ergibt sich der in folgender Grafik dargestellte Zusammenhang.

```python
"""Zielgröße"""

Teta = np.linspace(1100, 1301, 200) 
Zeit = (77-b1-b2*Teta)/(b3+b4*Teta)

TetaVsraum=np.linspace(1175, 1300, 124) 

ZeitVsraum= (77-b1-b2*TetaVsraum)/(b3+b4*TetaVsraum)

plt.plot(Teta, Zeit)
plt.xlabel('Ofentemperatur °C')
plt.ylabel('Verweilzeit  Min')
plt.title('Parameterkombination')
plt.fill_between(Teta, 0, 10, where=Teta>((77-b1)/b2), color='y')
plt.annotate('Kombinationen im Versuchsraum', (1175, 0.0), (1190, -5) )
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.ylim(top=20)
```

Da die Verweilzeit in dem Parameterraum liegen muss, werden die Parameterkombinationen auf den blauen Bereich eingeschränkt.

d) Um eine gute Reproduzierbarkeit des Prozesses zu bekommen, muss die Standardabwei-chung des Fertigungsprozesses minimiert werden. Sie ergibt sich maßgeblich aus der variablen Öltemperatur. Die durch die Öltemperatur hervorgerufene Standardabweichung wird für jeden Versuchspunkt bestimmt und über eine Regressionsfunktion abgeschätzt.

$Sig = {b_0} + {b_1} \cdot {\vartheta _{OFEN}} + {b_2} \cdot {T_{OFEN}} + {b_3} \cdot {\vartheta _{OFEN}} \cdot {T_{OFEN}}$

```python
'''Berechnung der Standardabweichung an jedem Versuchspunkt '''

EMatrix = np.reshape(Elas,(4,6))
StdElas=EMatrix.std(ddof=1,axis=1)

'''Regressionsmodell '''

data_Dic_Std= {'X':[1100, 1100, 1300, 1300], 'Y':[2, 10, 2, 10], 'Z':StdElas}

model1 = ols("Z ~ X + Y + X*Y " , data_Dic_Std).fit()
print(model1.summary())

RegStd=model1.predict({'X':TetaVsraum , 'Y':ZeitVsraum})


plt.plot(TetaVsraum, RegStd)
plt.xlabel('Ofentemperatur °C')
plt.ylabel('Standardabweichung KN/mm2 ')
plt.title('Standardabweichung für Parameterkombinationen')
plt.grid(True, which= 'both', axis='both', linestyle='--')
```

Es ergibt sich b0 = 6.8623, b1 = - 0.0029, b2 = - 0.9930 und b3 = 0.0006. Für die Parameter-kombinationen aus Aufgabenteil c) ergibt sich eine Standardabweichung als Funktion der Ofen-Temperatur. Sie ist in dem Bild oben bereits grafisch dargestellt. 

```python
TetaVsraum[np.argmin(RegStd)], ZeitVsraum[np.argmin(RegStd)]
```

Die Standardabweichung ist am kleinsten, wenn der Betriebspunkt mit ${\vartheta _{OFEN}}$= 1300 °C und ${T_{OFEN}}$ = 8.1656 Min. eingehalten wird.
