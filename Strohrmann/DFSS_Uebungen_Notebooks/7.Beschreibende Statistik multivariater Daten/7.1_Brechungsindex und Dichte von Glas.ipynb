## 7.1 Brechungsindex und Dichte von Glas




a) Die Stichprobe kann mit folgender Programmsequenz als Streudiagramm dargestellt werden.


Es ergibt sich folgendes Streudiagramm.


```python
""" Bibliotheken importieren"""
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

'''Laden der Daten'''
values= loadmat('Glasuntersuchung')['values']
n = values[:,0]
roh = values[:,1]


'''Darstellung des Streudiagramms'''
plt.scatter(n,roh, color='b')
plt.grid('True')
plt.xlabel('Brechungsindex n')
plt.ylabel('Dichte rho g/cm$^3$')
plt.xlim(right=1.525, left=1.513)
plt.ylim(top=2.52, bottom=2.47)
```

b) Die kumulativen Randhäufigkeiten ergeben sich durch Sortieren des jeweiligen Datenvektors. Die Wahrscheinlichkeit p$_0$ beträgt bei einem Stichprobenumfang N = 10 für jeden einzelnen Wert

${p_0} = \frac{1}{{10}} = 0.1$

Damit ergeben sich die im folgenden Bild dargestellten kumulativen Randhäufigkeiten.

Es wurde folgender Python-Code eingesetzt.


```python
''' Kumulative Randhäufigkeiten '''
nsort = np.append(np.sort(n),0)
rohsort =np.append(np.sort(roh),0)
H =np.append(0,np.cumsum(1/len(n)*np.ones(np.shape(n))))


plt.step(nsort,H, color='b')
plt.xlim(right=1.5236, left=1.513)
plt.ylim(top=1, bottom=0)
plt.grid('True')
plt.xlabel('Brechungsindex n')
plt.ylabel('Kumulative Randhäufigkeit H$_n$(n)')
plt.title('Brechungsindex')
```

```python
plt.step(rohsort,H, color='b')
plt.xlim(right=2.52, left=2.47)
plt.ylim(top=1, bottom=0)
plt.grid('True')
plt.xlabel('Dichte \rho g/cm^3')
plt.ylabel('Kumulative Randhäufigkeit H_{ \rho}(\rho)')
plt.title('Dichte')
```

c) Die Stichprobe wird über den Mittelwertsvektor




charakterisiert. Auf der Hauptdiagnonalen sind die Varianzen von Brechungsindex n und Dichte $\rho$, die beiden übrigen Elemente der Matrix geben die Kovarianz der beiden Größen an. Die Werte wurden mit folgender Python-Sequenz berechnet.





```python
'''Berechung von Mittelwert und Kovarianzmatrix'''
np.mean(values, axis=0), np.cov(np.transpose(values))
```
