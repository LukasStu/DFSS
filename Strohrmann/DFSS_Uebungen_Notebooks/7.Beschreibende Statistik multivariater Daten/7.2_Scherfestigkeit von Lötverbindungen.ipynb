### 7.2	Scherfestigkeit von LÃ¶tverbindungen


a) Die Stichprobe kann als Streudiagramm dargestellt werden.



Es wurde folgender Python-Code verwendet:








```python
""" Bibliotheken importieren"""
from scipy.io import loadmat
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  

'''Laden der Daten'''
values= loadmat('LoetenScherfestigkeit')['values']
T = values[:,0]
N = values[:,1]
F = values[:,2]
```

```python
f1, axes1 = plt.subplots(3, 3)

axes1[0, 0].hist(T, [125,150, 155], width=4, color='b', weights=(T/T)/(len(T)))
axes1[0, 0].grid(True, ls='--')
axes1[0, 0].axis([100, 175, 0, 0.8])

axes1[0, 1].scatter(N,T, color='b')
axes1[0, 1].grid(True, ls='--')
axes1[0, 1].axis([-125, 1125, 100, 175])

axes1[0, 2].scatter(F,T, color='b')
axes1[0, 2].grid(True, ls='--')
axes1[0, 2].axis([20, 80, 100, 175])

axes1[1, 0].scatter(T,N, color='b')
axes1[1, 0].grid(True, ls='--')
axes1[1, 0].axis([100, 175, 0, 1000])

axes1[1, 1].hist(N,np.arange(0,2000, 250), width=50, weights=(T/T)/(len(N)), color='b')
axes1[1, 1].grid(True, ls='--')
axes1[1, 1].axis([-250, 1500, 0, 0.6])

axes1[1, 2].scatter(F,N,color='b')
axes1[1, 2].grid(True, ls='--')
axes1[1, 2].axis([20, 80, 0, 1000])

axes1[2, 0].scatter(T,F, color='b')
axes1[2, 0].grid(True, ls='--')
axes1[2, 0].axis([100, 175, 20, 80])

axes1[2, 1].scatter(N,F,color='b')
axes1[2, 1].grid(True, ls='--')
axes1[2, 1].axis([0, 1000, 20, 80])

axes1[2, 2].hist(F,np.arange(30,90, 10),width=2,weights=(F/F)/(len(F)), color='b')
axes1[2, 2].grid(True, ls='--')
axes1[2, 2].axis([20, 80, 0, 0.4])
```

b) Die kumulativen RandhÃ¤ufigkeiten werden mit folgender Python-Sequenz berechnet:


```python
'''Kumulative RandhÃ¤ufigkeiten''' 
Tsort = np.append(0,np.sort(T))
Tsort = np.append(Tsort,150)

Nsort = np.append(0,np.sort(N))
Nsort = np.append(Nsort, 1000)

Fsort = np.append(0,np.sort(F))
Fsort = np.append(Fsort, 74.5)

H =np.append([0,0],np.cumsum(1/len(T)*np.ones(np.shape(T))))
```

Sie kÃ¶nnen als Liniendiagramm dargestellt werden.

```python
f2, axes2 = plt.subplots(1, 3)

axes2[0].step(Tsort,H, color='b')
axes2[0].grid(True, ls='--')
axes2[0].set_xlabel('$\Delta$T / Â°C')
axes2[0].set_ylabel('Kumulative RandhÃ¤ufigkeit H$_T$(T)')
axes2[0].set_title('Temperaturhub')

axes2[1].step(Nsort, H,color='b')
axes2[1].grid(True, ls='--')
axes2[1].set_xlabel('Anzahl')
axes2[1].set_ylabel('Kumulative RandhÃ¤ufigkeit H$_N$(N)')
axes2[1].set_title('Temperaturzyklen')

axes2[2].step(Fsort, H,color='b')
axes2[2].grid(True, ls='--')
axes2[2].set_xlabel('Scherfestigkeit / N')
axes2[2].set_ylabel('Kumulative RandhÃ¤ufigkeit H$_F$(F)')
axes2[2].set_title('Scherfestigkeit')
```
