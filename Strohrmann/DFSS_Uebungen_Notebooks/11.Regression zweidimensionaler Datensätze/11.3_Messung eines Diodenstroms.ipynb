## 11	Regression zweidimensionaler Datensätze
# 11.3	Messung eines Diodenstroms

a) Der Datensatz wird zunächst eingelesen


```python
import numpy as np
from scipy import io
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.formula.api import ols
import Pred_Band
%matplotlib inline

""" Laden der Daten"""
data =io.loadmat('Diodenstrom')

id=data['id'].reshape(16)
ud=data['ud'].reshape(16)
```

Anschließend wird ein Regressionsmodell aufgebaut. 

```python
''' Statistische Bewertung '''
'''Model [0; 1; 2;3],  ein kubisches Regressionsmodell  '''
# Model: id=b0+b1*ud+b2*ud**2+b3*ud**3

data_Dic={'X':ud, 'Y':ud**2 ,'E':ud**3, 'Z':id}

model1 = ols("Z ~ X + Y + E", data_Dic).fit()
print(model1.summary())
beta1=list(model1._results.params)
a,b,c,d=beta1
beta1.reverse()
```

b) Die Signifikanzbewertung (P_values) zeigt, dass keine Terme entfernt werden können. 

c) Zur Darstellung des Prognoseintervalls für zukünftige Messwerte mit $\gamma$  = 99 %verwendet, da alle Terme bis zur Ordnung 3 

Es ergeben sich folgende Abbildungenvorkommen.

```python
id_reg = np.polyval(beta1, ud)

""" grafische Ausgabe der Messpunkte und der Regressionsfunktion"""

plt.scatter(ud, id, c="g",  linewidth=0.5)
plt.plot(ud, id_reg, c="r", linestyle='--', linewidth=1)

plt.xlabel('Diodenspannung ${U_D}$ / V')
plt.ylabel('Diodenstrom ${I_D}$ / mA')
plt.grid(True, which= 'both', axis='both', linestyle='--')



def f(ud ,a,b,c,d):
    return a+ b*ud+c*ud**2+d*ud**3 

lpb, upb = Pred_Band.predband(ud, ud, id, [a,b,c,d], f, conf=0.95)
plt.plot(ud,lpb, 'b', linewidth= 0.3)
plt.plot(ud,upb, 'b', linewidth= 0.3)
plt.legend(['regression', 'Oberer Prognosebereich', 'Unterer Prognosebereich','Messung', ' Konfidenzbereich'])

```

```python
%matplotlib inline
"""Berechnung der Residuen"""

res1 = id_reg-id

"""grafische Ausgabe der Residuen"""
fig, ax = plt.subplots()
plt.stem(ud,res1,use_line_collection=True)
plt.xlabel('Diodenspannung ${U_D}$ / V')
plt.ylabel('Diodenstrom ${I_D}$ / mA')
plt.grid(True, which= 'both', axis='both', linestyle='--')
```

d) Zur Bestimmung der Regressionsfunktion von dem Zusammenhang zwischen Diodenspan-nung uD und dem Logarithmus des Diodenstroms iD wird identisch vorgegangen.

```python
'''Regression des Logarithmus von dem Diodenstrom'''
''' Statistische Bewertung '''
'''Model [0; 1; 2;3],  ein kubisches Regressionsmodell  '''
# Model: log(id)=b0+b1*ud+b2*ud**2+b3*ud**3

data_Dic={'X':ud, 'Y':ud**2 ,'E':ud**3, 'Z':np.log(id)}

model2_1 = ols("Z ~ X + Y + E", data_Dic).fit()
print(model2_1.summary())

```

###### e) Die Analyse der Terme auf Signifikanz zeigt, dass alle Terme bis auf den konstanten und linearen Term entfernt werden können.

```python
'''Regression des Logarithmus von dem Diodenstrom'''
''' Statistische Bewertung '''
'''Model [0; 1; 2],  Modellreduktion um kubischen Term  '''
# Model: log(id)=b0+b1*ud+b2*ud**2

data_Dic={'X':ud, 'Y':ud**2 , 'Z':np.log(id)}

model2_2 = ols("Z ~ X + Y ", data_Dic).fit()
print(model2_2.summary())
```

```python
'''Regression des Logarithmus von dem Diodenstrom'''
''' Statistische Bewertung '''
'''Model [0; 1],  Modellreduktion um quadratischen Term  '''
# Model: log(id)=b0+b1*ud

data_Dic={'X':ud, 'Z':np.log(id)}

model2_3 = ols("Z ~ X ", data_Dic).fit()
print(model2_3.summary())
beta2=list(model2_3._results.params)
a,b=beta2
beta2.reverse()
```

###### Beide Terme sind signifikant, keine weitere Reduktion von Termen möglich.



Zur Bestimmung des Diodenstroms ${i_D}$ muss die Exponentialfunktion der über die Regressionsfunktion berechneten Werte bestimmt werden.

```python
'''Berechnung der Funktionswerte wieder mit polyval und Pred_Band.predband'''
id_log_reg = np.polyval(beta2, ud)


""" grafische Ausgabe der Messpunkte und der Regressionsfunktion"""

plt.scatter(ud, id, c="g",  linewidth=0.5)
plt.plot(ud, np.exp(id_log_reg), c="r", linestyle='--', linewidth=1)

plt.xlabel('Diodenspannung ${U_D}$ / V')
plt.ylabel('Diodenstrom ${I_D}$ / mA')
plt.grid(True, which= 'both', axis='both', linestyle='--')


def f(ud ,a,b):
    return a+ b*ud 

lpb, upb = Pred_Band.predband(ud, ud, np.log(id), [a,b], f, conf=0.95)
plt.plot(ud,np.exp(lpb), 'b', linewidth= 0.3)
plt.plot(ud,np.exp(upb), 'b', linewidth= 0.3)
plt.legend(['regression', 'Oberer Prognosebereich', 'Unterer Prognosebereich','Messung', ' Konfidenzbereich'])
```

```python
"""Berechnung der Residuen"""

res2 = np.exp(id_log_reg)-id

"""grafische Ausgabe der Residuen"""
fig, ax = plt.subplots()
plt.stem(ud,res2,use_line_collection=True)
plt.xlabel('Diodenspannung ${U_D}$ / V')
plt.ylabel('Diodenstrom ${I_D}$ / mA')
plt.grid(True, which= 'both', axis='both', linestyle='--')
```

```python
''' Berechnung der Fehlerquadrate'''
a1 = np.sum((res1)**2)
a2 = np.sum((res2)**2)
a1,a2
```

g) Die Summe der Fehlerquadrate ist im Falle der Transformation (A = 2.5) kleiner als im Fall ohne Transformation (A = 3). Das liegt daran, dass die Funktion nach der Transformation praktisch eine Gerade ist, die sehr gut approximiert werden kann.
Allerdings ergeben sich im Fall der Transformation bei großen Diodenspannungen sehr große Prognoseintervalle bis zu 10 mA. Sie sind deutlich größer als im nicht transformierten Fall. 



Außerdem sind die ursprünglich normalverteilten Messfehler nach der Transformation der Messwerte mit der Logarithmusfunktion nicht mehr normalverteilt. Deshalb ist eine Annahme, die bei der Berechnung der Prognoseintervalle gemacht wird, verletzt.
