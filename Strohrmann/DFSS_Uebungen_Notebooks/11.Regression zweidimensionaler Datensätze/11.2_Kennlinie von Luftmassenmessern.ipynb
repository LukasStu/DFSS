a) Der Datensatz wird zunächst eingelesen

```python
import numpy as np
from scipy import io
import matplotlib.pyplot as plt
from scipy import stats
from statsmodels.formula.api import ols
%matplotlib notebook

""" Laden der Daten"""
data =io.loadmat('Messwerte')['Messwerte']

m=data[0,:]
U=data[1,:]
m.shape
```

#### Anschließend wird ein Regressionsmodell aufgebaut.







```python
''' Statistische Bewertung '''
'''Model [0; 1; 2;3],  ein kubisches Regressionsmodell  '''
# Model: U=b0+b1*m+b2*m**2+b3*m**3

data_Dic={'X':m, 'Y':m**2 ,'E':m**3, 'Z':U}

model = ols("Z ~ X + Y + E", data_Dic).fit()
print(model.summary())
```

#### ???????? b) Die Signifikanzbewertung zeigt, dass die Konstante entfernt werden kann. Das adjungierte Bestimmtheitsmaß reduziert sich unwesentlich von 0.9887 auf 0.9883. Weitere Terme können nicht entfernt werden.

```python
''' Statistische Bewertung '''
'''Model [ 1; 2;3],  Modellreduktion um Konstante  '''
# Model: U=b1*m+b2*m**2+b3*m**3

data_Dic={'X':m, 'Y':m**2 ,'E':m**3, 'Z':U}

model1 = ols("Z ~ 0+X + Y + E", data_Dic).fit()
print(model1.summary())
beta1=list(model1._results.params)
beta1.reverse()
beta1.append(0)
```

c) Zur Bestimmung der Regressionsfunktion von dem inversen Zusammenhang müssen die beiden Variablem m und U ausgetauscht werden.

```python
''' Statistische Bewertung '''
'''Model [0; 1; 2;3],  ein kubisches Regressionsmodell  '''
# Model: m=b0+b1*U+b2*U**2+b3*U**3

data_Dic={'X':U, 'Y':U**2 ,'E':U**3, 'Z':m}

model = ols("Z ~ X + Y + E", data_Dic).fit()
print(model.summary())

```

###### ????? d) Die Analyse der Terme auf Signifikanz zeigt, dass die alle Terme nicht signifikant sind, außer dem quadratischen Term. Das Bestimmtheitsmaß steigt durch die Reduktion der Regressionsfunktion von 0.9991 auf 0.9993 geringfügig an.



```python
''' Statistische Bewertung '''
'''Model [1; 2;3],  Modellreduktion um Konstante  '''
# Model: m=b1*U+b2*U**2+b3*U**3

data_Dic={'X':U, 'Y':U**2 ,'E':U**3, 'Z':m}

model = ols("Z ~ 0+ X + Y + E", data_Dic).fit()
print(model.summary())
```

```python
''' Statistische Bewertung '''
'''Model [ 2;3],  Modellreduktion um linearen Term  '''
# Model: m=b2*U**2+b3*U**3

data_Dic={'Y':U**2 ,'E':U**3, 'Z':m}

model = ols("Z ~ 0 + Y + E", data_Dic).fit()
print(model.summary())
```

```python
''' Statistische Bewertung '''
'''Model [ 2;3],  Modellreduktion um kubischen Term  '''
# Model: m=b2*U**2

data_Dic={'Y':U**2 , 'Z':m}

model2 = ols("Z ~ 0 + Y ", data_Dic).fit()
print(model2.summary())

beta2=list(model2._results.params)
#beta2.append(0)
#beta2.insert(0,0)
#beta2.insert(0,0)
#m_reg = np.polyval(beta2, U)
beta2[0]
```

Der Regressionskoeffizient ist beta2 ist 18.1703. Damit ergibt sich die Gleichung


$m = 18.1703 \cdot {U^2}$
 
Auflösen nach U ergibt:


$U = \sqrt {\frac{m}{{18.1703}}}$


 




e) Nach Auflösen der Gleichung können alle die Stichprobenwerte und die beiden Regressionsfunktionen dargestellt werden.

```python
'''Berechnung der Funktionswerte'''

U1_reg = model1._results.params[0]*m +model1._results.params[1]*m**2 + model1._results.params[2]*m**3
#U1_reg = np.polyval(beta1, m)
U2_reg = np.sqrt(m/beta2[0])

""" grafische Ausgabe der Messpunkte und der Regressionsfunktion"""

plt.scatter(m, U, c="g",  linewidth=1)
plt.plot(m, U1_reg, c="r", linestyle='--', linewidth=1)
plt.plot(m, U2_reg, c="b", linestyle='--', linewidth=1)
plt.xlabel('Luftmassenstroms m')
plt.ylabel('Gangsspannung U')
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.legend([ 'Regression1', 'Regression2','Stichprobe'])
```

f) Die Residuen errechnen sich aus der Differenz von Stichprobenwert und Regressionswert.

```python
"""Berechnung der Residuen"""

res1 =  U1_reg-U
res2= U2_reg-U


""" Darstellung der Residuen"""
fig, ax = plt.subplots()
plt.stem(m,res1, use_line_collection=True,linefmt='--', markerfmt='D')
plt.stem(m,res2, use_line_collection=True)
plt.xlabel('Luftmassenstroms m')
plt.ylabel('Residuen Gangsspannung ')
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.legend([ 'Regression1', 'Regression2'])
```

```python
''' Berechnung der Fehlerquadrate'''
a1 = np.sum((res1)**2)
a2 = np.sum((res2)**2)
a1,a2
```

Für diesen direkten Vergleich der beiden Regressionsfunktionen wird die Summe der Fehlerquadrate berechnet. Bei der ersten Regression ergibt sich a1 = 0.1928 und bei der zweiten Regressionsfunktion ergibt sich a2 = 0.0099. Damit ist in diesem Fall die Regression der inversen Funktion erheblich genauer als bei der Funktion selbst. Das Ergebnis wird auch grafisch bestätigt.
