# 11 Regression zweidimensionaler Datensätze
# 11.1 Scherfestigkeit von Schweißverbindungen
a) Nach dem importieren der benötigten Bibliotheken werden die Daten eingelesen. Mithilfe des Befehls polyfit wird ein Regressionspolynom zweiten Grades berechnet und grafisch dargestellt.

```python
import numpy as np
import seaborn as sns
import pandas as pd
from scipy import io
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy import stats
from statsmodels.formula.api import ols
import Pred_Band
%matplotlib inline

""" Laden der Daten"""
data =io.loadmat('ScherfestigkeitSchweissen')['data']

S=data[:,0]
D=data[:,1]
```

```python
"""Erzeugen der Regressionsfunktion"""

[a,b,c] = np.polyfit(D, S, 2)
S_reg = np.polyval([a,b,c], D)


""" grafische Ausgabe der Messpunkte und der Regressionsfunktion"""

plt.scatter(D, S, c="b")
plt.plot(D, S_reg, c="r")
plt.xlabel('Durchmesser')
plt.ylabel('Scherfestigkeit')
plt.grid(True, which= 'both', axis='both', linestyle='--')
```

b) Die Residuen ergeben sich aus der Differenz von Datensatz und Regression an den entsprechenden Stellen. Sie können direkt in ein Diagramm eingezeichnet werden. Es ergibt sich die Abbildung da unten.




Es ist keine Struktur der Reststreuung zu erkennen, die auf einen systematischen Regressionsfehler schließen lässt.

```python
"""Berechnung der Residuen"""

res = S - S_reg


"""grafische Ausgabe der Residuen"""
fig, ax = plt.subplots()
plt.stem(D,res, use_line_collection=True)
plt.xlabel('Durchmesser')
plt.ylabel('Residuen Scherfestigkeit')
plt.grid(True, which= 'both', axis='both', linestyle='--')
```

c) Die Regression wird auf Basis der Reststreuung, des adjungierten Bestimmtheitsmaßes sowie der Signifikanz der einzelnen Regressionskoeffizienten bewertet. Das adjungierte Bestimmtheitsmaß ist mit RADJ = 0.9999 =1 sehr hoch. Offensichtlich wird die Streuung des Datensatzes mit der Regression gut abgebildet. Die Reststreuungsanalyse hat ebenfalls gezeigt, dass die Regressionsfunktion sinnvoll gewählt ist. Der t-Test zeigt, dass nicht alle Koeffizienten signifikant sind. Deshalb wird in den folgenden Aufgabenteilen eine Reduktion der Regressionsterme durchgeführt.

```python
''' Statistische Bewertung '''
'''Model [0; 1; 2], Vollquadratisch '''
# Model: z=a+bX+cY=a+bD+cD*D
data_Dic={'X':D, 'Y':D**2 , 'Z':S}

model = ols("Z ~ X + Y", data_Dic).fit()


print(model.summary())
print(model._results.params)

''' b2 = 0.5305 > 0.05 -> modell vereinfachen zu [0;1]'''
# Model: z=a+bX=a+bD
data_Dic={'X':D, 'Z':S}

model = ols("Z ~ X ", data_Dic).fit()


print(model.summary())
print(model._results.params)
model
```

##### d) Ausgehend von dem vollquadratischen Modell werden einzelne Regressionsterme eliminiert. Dabei ist wesentlich, dass nach jedem Streichen eines Regressionsterms eine erneute Bewertung der Signifikanz durchgeführt wird.


Es zeigt sich, dass nur die Koeffizienten   und   signifikant sind. Das adjungierte Be-stimmtheitsmaß und die P-Werte der einzelnen Koeffizienten sind in folgender Tabelle zusammengefasst. 



Durch die Eliminierung der nicht signifikanten Terme verschlechtert sich das adjungierte Bestimmtheitsmaß nicht. 

```python
[b,a] = np.polyfit(D, S, 1)
S_reg = np.polyval([b,a], D)


""" grafische Ausgabe der Messpunkte und der Regressionsfunktion"""

plt.scatter(D, S, c="b")
plt.plot(D, S_reg, c="r")
plt.xlabel('Durchmesser')
plt.ylabel('Scherfestigkeit')
plt.grid(True, which= 'both', axis='both', linestyle='--')
```
