# 8	Multivariate Wahrscheinlichkeitstheorie
### 8.1	Reihenschaltung von Widerständen



a) Der Bereich, für den die Wahrscheinlichkeit P$_0$ gilt, hat in der R$_1$ Dimension eine Breite von 100 $\Omega $, in der Dimension R$_2$ eine Breite von 200 $\Omega$. Da die Wahrscheinlichkeit für das sichere Ereignis 1 ist, gilt


$1 = \int\limits_{950}^{1050} {\int\limits_{2400}^{2600} {{P_0}\,\,d{R_2}\,d{R_1}} }  = 100 \cdot 200 \cdot {P_0} = 20000 \cdot {P_0}$


Damit ergibt sich die Wahrscheinlichkeit zu


${P_0} = \frac{1}{{20000}} = 50 \cdot {10^{ - 6}}$



b) Die Wahrscheinlichkeitsdichte ist in dem ausgezeichneten Bereich konstant. Es ergibt sich folgendes Diagramm.


```python
""" Bibliotheken importieren"""
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D

''' Definition der Datensaetze'''
r1 = np.array([900, 950, 950.001, 1050, 1050.001, 1100])
r2 = np.array([2300, 2400, 2400.001, 2600, 2600.001, 2700])
p0 = 1/20000
p1 = 1/100
p2 = 1/200
rc = np.array([3200, 3350, 3450, 3550, 3650, 3800])
rw = np.array([3200, 3449.999, 3450, 3550, 3550.001 ,3800])

'''Erzeugen der Matrizen für die Grafiken'''
R1,R2 = np.meshgrid(r1,r2)

P0 = p0*(R1>950)*(R1<=1050)*(R2>2400)*(R2<=2600) 

P0C = np.array(P0)
P0C[2,2] = 0
P0C[2,3:] = p0*100
P0C[3:,2] = p0*200
P0C[3:,3:] = 1
```

```python
'''Definition der Grafik'''
fig1 = plt.figure()
ax = Axes3D(fig1)
surf = ax.plot_surface(R1,R2,P0*1e6,cmap=cm.Blues_r)
ax.set_xlabel('Widerstand $R_1$ / $\Omega$')
ax.set_ylabel('Widerstand $R_2$ / $\Omega$')
ax.set_zlabel('F($R_1,R_2$)')
ax.set_title('Wahrscheinlichkeitsdichte')
```

```python
'''Definition der Grafik'''
fig2 = plt.figure()
ax = Axes3D(fig2)
surf = ax.plot_surface(R1,R2,P0C,cmap=cm.Blues_r)
ax.set_xlabel('Widerstand $R_1$ / $\Omega$')
ax.set_ylabel('Widerstand $R_2$ / $\Omega$')
ax.set_zlabel('F($R_1,R_2$)')
ax.set_title('Wahrscheinlichkeitsverteilung')
```

c) Die Verteilungsfunktion ergibt sich für den Bereich 950 $\Omega$ ≤ R$_1$ ≤ 1050 $\Omega$ und 
2400 $\Omega$ ≤ R$_2$ ≤ 2600 $\Omega$ aus dem Integral


$F\left( {{R_1},{R_2}} \right) = \int\limits_{950}^{{R_1}} {\int\limits_{2400}^{{R_2}} {{P_0}\,\,d{R_2}\,d{R_1}} }  = {P_0} \cdot \left( {{R_1} - 950} \right) \cdot \left( {{R_2} - 2400} \right)$


Es ist in dem Diagramm oben bereits darstellt.






d) Die Randverteilungen sind gleichverteilt. Danit kann die Wahrscheinlichkeitsdichte der Randverteilung von R1 im Bereich 950 $\Omega$ ≤ R1 ≤ 1050 $\Omega$ über das sichere Ereignis 


$1 = \int\limits_{950}^{1050} {{P_1}\,\,d{R_1}}  = 100 \cdot {P_1}$

berechnet werden zu

${P_1} = \frac{1}{{100}}$

Für $R_2$ ergibt sich im Bereich 2400 $\Omega$ ≤ R2 ≤ 2600 $\Omega$ analog

${P_2} = \frac{1}{{200}}$

```python
'''Erzeugen der Matrizen für die Grafiken'''
P1 = p1*(r1>950)*(r1<=1050)
P1C = np.array([0, 0, 0, 1, 1, 1])
P2 = p2*(r2>2400)*(r2<=2600)
P2C = np.array([0, 0 ,0 ,1, 1, 1])
PC = np.array([0, 0, 1, 1, 0, 0 ])*p1*p2*100

'''Definition der Grafik'''
plt.plot(R1[0,:],P1*1000,'b')
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Widerstand $R_1$ / $\Omega$')
plt.ylabel('F($R_1).1000$')
plt.title('Wahrscheinlichkeitsdichte $R_1$')
plt.ylim(top=15)
```

```python
'''Definition der Grafik'''
plt.plot(R2[:,0],P2*1000,'b')
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Widerstand $R_2$ / $\Omega$')
plt.ylabel('F($R_2).1000$')
plt.title('Wahrscheinlichkeitsdichte $R_2$')
plt.ylim(top=15)
```

e) Die kumulativen Randverteilung von R$_1$ ergibt sich über die Gleichung

${F_{{R_1}}}\left( {{R_1}} \right) = \int\limits_{950}^{{R_1}} {\int\limits_{2400}^{2600} {{P_0}\,\,d{R_2}\,d{R_1}} }  = {P_0} \cdot \left( {{R_1} - 950} \right) \cdot \left( {2600 - 2400} \right) = {P_0} \cdot 200 \cdot \left( {{R_1} - 950} \right)$

und für R$_2$ über 

${F_{{R_2}}}\left( {{R_2}} \right) = \int\limits_{950}^{1050} {\int\limits_{2400}^{{R_2}} {{P_0}\,\,d{R_2}\,d{R_1}} }  = {P_0} \cdot \left( {1050 - 950} \right) \cdot \left( {{R_2} - 2400} \right) = {P_0} \cdot 100 \cdot \left( {{R_2} - 2400} \right)$




```python
'''Definition der Grafik'''
plt.plot(R1[0,:],P1C,'b')
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Widerstand $R_1$ / $\Omega$')
plt.ylabel('F($R_1).1000$')
plt.title('Verteilungfunktion $R_1$')
```

```python
'''Definition der Grafik'''
plt.plot(R2[:,0],P1C,'b')
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Widerstand $R_2$ / $\Omega$')
plt.ylabel('F($R_2).1000$')
plt.title('Verteilungfunktion $R_2$')
```

f) Der angegebene Bereich liegt in Bereich konstanter Wahrscheinlichkeitsdichte. Damit ergibt sich die Wahrscheinlichkeit zu 


$P = \int\limits_{990}^{1010} {\int\limits_{2450}^{2550} {{P_0}\,\,d{R_2}\,d{R_1}} }  = {P_0} \cdot \left( {1010 - 990} \right) \cdot \left( {2550 - 2450} \right) = 0.1$



g) Damit zwei Zufallsvariablen unabhängig sind, muss die Bedingung 

$F\left( {{R_1},{R_2}} \right) = {F_{{R_1}}}\left( {{R_1}} \right) \cdot {F_{{R_2}}}\left( {{R_2}} \right)$


erfüllt sein. Nach den Rechnungen oben gilt

$F\left( {{R_1},{R_2}} \right) = {P_0} \cdot \left( {{R_1} - 950} \right) \cdot \left( {{R_2} - 2400} \right)$



und


$\begin{array}{c}{F_{{R_1}}}\left( {{R_1}} \right) \cdot {F_{{R_2}}}\left( {{R_2}} \right) = {P_0} \cdot 200 \cdot \left( {{R_1} - 950} \right) \cdot {P_0} \cdot 100 \cdot \left( {{R_2} - 2400} \right)\\ = \frac{1}{{20000}} \cdot 200 \cdot \left( {{R_1} - 950} \right) \cdot {P_0} \cdot 100 \cdot \left( {{R_2} - 2400} \right)\\ = \left( {{R_1} - 950} \right) \cdot {P_0} \cdot 100 \cdot \left( {{R_2} - 2400} \right)\end{array}$


Beide Werte stimmen überein, die beiden Zufallsvariablen R$_1$ und R$_2$ sind statistisch unabhängig.







h) Weil die beiden Zufallsvariablen unabhängig sind, ergibt sich die Wahrscheinlichkeitsdichte f(R) aus der Faltung der beiden Wahrscheinlichkeitsdichten f(R$_1$) und f(R$_2$).

$f\left( R \right) = {f_{{R_1}}}\left( {{R_1}} \right) \cdot {f_{{R_2}}}\left( {{R_2}} \right)$


Mithilfe der grafischen Faltung ergibt sich die folgende Wahrscheinlichkeitsdichte.



i) Die gesuchte Wahrscheinlichkeit entspricht der Fläche, die über die rote Linie begrenzt wird.

$P = 100 \cdot 5 \cdot {10^{ - 3}} = 0.5$

```python
'''Definition der Grafik'''
plt.plot(rc,PC*1e3,'b', linewidth=1)
plt.plot(rw,PC*1e3,'r',linewidth=1,linestyle='--')
plt.legend(['Aufgabenteil h','Aufgabenteil i'], loc=1)
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Widerstand $R$ / $\Omega$')
plt.ylabel('Wahrscheinlichkeitsdichte f (R).$10^3$')
plt.ylim(top=7)
```
