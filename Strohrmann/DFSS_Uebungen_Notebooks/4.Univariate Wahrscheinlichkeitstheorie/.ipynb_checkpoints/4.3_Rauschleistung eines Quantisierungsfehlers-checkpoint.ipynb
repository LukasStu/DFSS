# 4.Univariate Wahrscheinlichkeitstheorie


### 4.3	Rauschleistung eines Quantisierungsfehlers



a) Ein 8-Bit-Wandler hat bei einer Refernzspannung von UREF = 5 V eine Auflösung $\Delta$U von 


$\Delta U = \frac{{{U_{REF}}}}{{{2^8}}} = \frac{5}{{256}} = {\rm{0}}{\rm{.0195}}\,\,{\rm{V}}$

Der bei der Quatisierung entstehende Fehler liegt damit zwischen 0 und $\Delta$U. Er weist eine Gleichverteilung auf. Aufgrund der Normierung 

$\int\limits_0^{\Delta U} {f\left( u \right)\,\,du = 1} $

besitzt die Wahrscheinlichkeitsdichte f(u) in dem Intervall von 0 … $\Delta$U den konstanten Wert 
f(u) = 1/$\Delta$U = 1/(0.0195 V). Damit kann die Wahrscheinlichkeitsdichte grafisch dargestellt wer-den.


```python
import numpy as np
import matplotlib.pyplot  as plt

U=np.arange(-0.01,0.3,0.0001)
Udelta=5/2**8

WarUdelta=np.empty((len(U)))
for x in range (len(U)):
    if U[x]>0 and U[x]<= Udelta:
        WarUdelta[x]= 1/Udelta 
    else:
        WarUdelta[x]=0
        
plt.plot(U,WarUdelta, color='b')    
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Spannung u / V')
plt.ylabel('Wahrscheinlichkeisdichte f(u)  .V')
plt.xlim(left= -0.01, right=0.03)
```

b) Die Leistung eines Signals ist als das Quadrat des Signals definiert. Die mittlere Rauschleitung ergibt sich aus dem Erwartungswert von $u^2$. 

$E\left( {{u^2}} \right) = \int\limits_0^{\Delta U} {{u^2} \cdot f\left( u \right)\,\,du}  = \int\limits_0^{\Delta U} {{u^2} \cdot \frac{1}{{\Delta U}}\,\,du}  = \left. {{P_0} \cdot \frac{1}{3} \cdot {u^3}} \right|_o^{\Delta U} = \frac{1}{{\Delta U}} \cdot \frac{1}{3} \cdot \left( {\Delta {U^3} - 0} \right) = \frac{{\Delta {U^2}}}{3} = 0.{\rm{12675}}\,\,{\rm{mW}}$


c) Bei einem normalverteilten Signal ergibt sich 

${\sigma ^2} = E\left( {{{\left( {u - \mu } \right)}^2}} \right) = E\left( {{u^2} - 2 \cdot E\left( u \right) \cdot \mu  + {\mu ^2}} \right) = E\left( {{u^2} - {\mu ^2}} \right) = E\left( {{u^2}} \right) - {\mu ^2}$


Auflösen nach E($u^2$) führt zu

$E\left( {{u^2}} \right) = {\sigma ^2} + {\mu ^2}$

Die mittlere Leistung ergibt sich aus der Summe von quadriertem Mittelwert und Varianz.
