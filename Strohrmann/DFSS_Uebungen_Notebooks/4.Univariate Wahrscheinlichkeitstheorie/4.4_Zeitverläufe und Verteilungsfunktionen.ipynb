# 4.Univariate Wahrscheinlichkeitstheorie

### 4.4	Zeitverläufe und Verteilungsfunktionen

a) Das dargestellte Sägezahnsignal zeigt über der Zeit einen linearen Anstieg. Damit sind alle Werte zwischen - 1 und 1 gleich wahrscheinlich. Bei der gesuchten Verteilung handelt sich deshalb um eine Gleichverteilung zwischen - 1 und 1. Wahrscheinlichkeitsdichte und Verteilungs-funktions sind in den folgenden Diagrammen dargestellt. Aufgrund der Normierungsbedingung hat die Wahrscheinlichkeitsdichte in dem Intervall von - 1 bis 1 von f(x) = 1/2.





```python
import numpy as np
import matplotlib.pyplot  as plt

x = np.arange(-3,3,0.0001)
x = x + 1e-12
f = 1/2*(np.heaviside(x+1,0) - np.heaviside(x-1,0))
F = 1/2*(x+1)*(np.heaviside(x+1,0) - np.heaviside(x-1,0)) + np.heaviside(x-1,0)

plt.plot(x,f, color='b')    
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Zufallsvariable x')
plt.ylabel('f(x)')
plt.xlim(left= -1.5, right=1.5)
plt.title('Wahrscheinlichkeitsdichte F(x)')
```

```python
plt.plot(x,F, color='b')    
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Zufallsvariable x')
plt.ylabel('F(x)')
plt.xlim(left= -1.5, right=1.5)
plt.title('Verteilungsfunktion')
```

b) Die Berechnung der Wahrscheinlichkeitsdichte wird des Signals x(t) wird auf die Transforma-tion einer Zufallsvariablen zurückgeführt. Es wird von einer Variablen t ausgegangen, die in dem Intervall von 0 … 1 gleichverteilt ist. Die Wahrscheinlichkeitsdichte von t lautet in dem Intervall von 0 … 1

${f_T}\left( t \right) = 1$
und die entsprechende Verteilungsfunktion

${F_T}\left( t \right) = t$

Die Zufallsvariable t wird mit der Funktion 

$x\left( t \right) = \cos \left( {\frac{{2 \cdot \pi }}{2} \cdot t} \right) = \cos \left( {\pi  \cdot t} \right)$

auf die Zufallsvariable x abgebildet. In den Zahlenbereich 0 ≤ t ≤ 1 ist die Funktion streng mo-noton fallend. Deshalb lautet die Umkehrfunktion in dem Bereich von 0 < t ≤ 1 

$t = \frac{1}{\pi } \cdot \arccos \left( x \right)$

Die Verteilungsfunktion der Variable x errechnet sich damit zu

${F_X}\left( x \right) = {F_T}\left( {\frac{1}{\pi } \cdot \arccos \left( x \right)} \right) = {\left. t \right|_{t = \frac{1}{\pi } \cdot \arccos \left( x \right)}} = \frac{1}{\pi } \cdot \arccos \left( x \right)$


Die Wahrscheinlichkeitsdichte errechnet sich über 


${f_X}\left( x \right) = {f_T}\left( {\frac{1}{\pi } \cdot \arccos \left( x \right)} \right) \cdot \left| {\frac{1}{\pi } \cdot \frac{{d\,\,\arccos \left( x \right)}}{{dx}}} \right| = \frac{1}{\pi } \cdot \frac{1}{{\sqrt {1 - {x^2}} }}$

Wahrscheinlichkeitsdichte und Verteilungsfunktion der Zufallsvariable x sind in den folgenden Diagrammen dargestellt.

```python
x = np.arange(-1,1,0.0001)
x = x + 1e-12

f2 = (1/np.pi)*(1/np.sqrt(1-x**2))*(np.heaviside(x+1,0) - np.heaviside(x-1,0))
F2 = ((np.arccos(-x))*(1/np.pi)*(np.heaviside(x+1,0) - np.heaviside(x-1,0))) + np.heaviside(x-1,0)
```

```python
plt.plot(x,f2, color='b')    
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Zufallsvariable x')
plt.ylabel('F(x)')
plt.xlim(left= -2, right=2)
plt.title('Wahrscheinlichkeitsdichte F(x)')

plt.ylim(bottom=-0.5, top=10)
```

```python
plt.plot(x,F2, color='b')    
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.xlabel('Zufallsvariable x')
plt.ylabel('F(x)')
plt.xlim(left= -2, right=2)
plt.title('Verteilungsfunktion')

plt.ylim(bottom=-0.1, top=1.1)
```
