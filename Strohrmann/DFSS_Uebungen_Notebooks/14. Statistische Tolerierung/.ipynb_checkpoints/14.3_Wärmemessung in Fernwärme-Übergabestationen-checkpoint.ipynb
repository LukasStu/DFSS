# 14.3	Wrmemessung in Fernwrme-bergabestationen



```python
import numpy as np
import sympy as syms
from sympy.core import evalf
from scipy.stats import norm
from scipy.stats import uniform
import matplotlib.pyplot as plt
%matplotlib inline  


#Definition der Kenngren
T0 = 25
T1 = 25
T2 = 70
sigT12 = 10e-3
alpha = 4e-3
R0 = 100
R1 = 100*(1+alpha*(T1-T0))
R2 = 100*(1+alpha*(T2-T0))
sigR = 1e-3


#Temperaturabhngigkeit des Widerstands
T = np.arange(0,101,1)
R = 100*(1+alpha*(T-T0))
 
#berprfen der berechneten Temperatur
Tref = T0 + (R*(T1-T2)-R2*(T1-T0)+R1*(T2-T0))/(R1-R2)
```

```python
#Berechnen der Empfindlichkeiten
ET1 = (R-R2)/(R1-R2)
ET2 = (R1-R)/(R1-R2)
ER = (T1-T2)/(R1-R2)
ER1 = (R2-R)*(T1-T2)/(R1-R2)**2
ER2 = (R-R1)*(T1-T2)/(R1-R2)**2

#Berechnen der Standardabweichung 
sigT = ((ER1**2+ER2**2+ER**2)*sigR**2 + (ET1**2+ET2**2)*sigT12**2)**0.5
```


c) Der Messfehler ist von der Temperatur abhngig, weil die Empfindlichkeiten vom Widerstand R und damit von der Temperatur T abhngen. Das wurde bei der Programmierung oben bereits bercksichtigt, indem T und R als Vektoren definiert sind.



d) Der Messfehler ist in folgendem Diagramm als $2.\sigma_T$ Grenzen eingezeichnet. Zustzlich ist die unter e) geforderte Monte-Carlo-Simulation und im Bild links die unter a) berechnete Temperatur als Funktion der realen Temperatur dargestellt .





e) Bei der Monte-Carlo-Simulation werden fr die Gren R1, R2, T1 und T2 Zufallszahlen generiert. Sie sind alle normalverteilt und besitzen die angegebenen Sollwerte. Die Gre T0 ist fest definiert. Die Gre R entspricht ihrem Sollwert als Funktion der Temperatur und besitzt die angegebenen Standardabweichung. Die Zufallszahlen werden in die Formel 



### $T=T_0+\frac{R(T_1-T_2 )-R_2(T_1-T_0 )+R_1(T_2-T_0 )}{R_1-R_2}$


eingesetzt. Die bereits oben gezeigten Simulationswerte besttigen die berechnete Toleranz.

```python
# Monte Carlo Simulation, Generieren der Zufallsgren
N = 10000
T0V = T0*np.ones((N,1))
T1V =norm.rvs(T1,sigT12,(N,1))
T2V = norm.rvs(T2,sigT12,(N,1))
R1V = norm.rvs(R1,sigR,(N,1))
R2V = norm.rvs(R2,sigR,(N,1))
TV =uniform.rvs(0,100,(N,1))

RV = 100*(1+alpha*(TV-T0)) + norm.rvs(0,sigR,(N,1))
TrefV = T0V + (RV*(T1V-T2V)-R2V*(T1V-T0V)+R1V*(T2V-T0V))/(R1V-R2V)
```

```python
plt.scatter(TV,abs(TrefV-TV),marker='+')

plt.plot(T,2*sigT,'r')

plt.ylabel('Toleranz / C')
plt.xlabel('Temperatur / C')
plt.legend(('Grenze $2\cdot\sigma_T$','Monte-Carlo          '),loc='upper center')
plt.grid()
```

```python
plt.plot(T,Tref,'b')
plt.xlabel('Soll-Temperatur / C')  
plt.ylabel('Berechnete Temperatur / C')  
plt.grid()
```
