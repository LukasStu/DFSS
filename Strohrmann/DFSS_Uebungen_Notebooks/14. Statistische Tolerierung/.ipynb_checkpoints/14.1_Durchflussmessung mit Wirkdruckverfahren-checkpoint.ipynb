# 14	Statistische Tolerierung
## 14.1	Durchflussmessung mit Wirkdruckverfahren

a) Zur Herleitung wird vom Analog-Digital-Wandler ausgegangen, Referenzspannung und Mess-größe werden eingesetzt. Auflösen der Gleichung in die Summanden ergibt 

$$ N=\frac{(m⋅p⋅U_{REF}+U_A)}{U_{ADC}} =\frac{(m⋅Q^2/k^2 ⋅(U_{ADC}+U_{OFF} )+U_A)}{U_{ADC}} =m⋅\frac{Q^2}{k^2} +m⋅\frac{Q^2}{k^2} ⋅\frac{U_{OFF}}{U_{ADC}} +N_A $$


Definition der Größen, Berechnung der Kennlinie und Darstellung in Python



```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
""" Definition der Sollwerte und Toleranzen"""
k = 0.2
m = 0.04
UADC0 = 5
dUADC = 500e-3
sigUADC = dUADC/6
Uoff0 = 10e-3
dUoff = 20e-3
sigUoff = dUoff/12**0.5
NA0 = 0
dNA = 2/2048
sigNA = dNA/12**0.5
```

```python
""" Berechnung der Kennlinie"""
Q= np.linspace(0,1,10)

#N0 = m*Q.^2/k^2*(UADC0+Uoff0)+NA0;

N0 = (m*(Q**2)/k**2*(UADC0+Uoff0)+NA0)/5

plt.plot(Q, N0, c="b")
plt.xlabel('Durchfluss / m³/h')
plt.ylabel('Normierter ADC-Wert N')
plt.grid(True, which= 'both', axis='both', linestyle='--')
plt.show
```

b) Die linearisierte Maßkette ergibt sich aus dem vollständigen Fehlerdifferential


$ΔN=\frac{∂N}{(∂U_{ADC})}⋅ΔU_{ADC}+\frac{∂N}{(∂U_{OFF})}⋅ΔU_{OFF}+\frac{∂N}{(∂N_{A})}⋅ΔN_{A}$


mit


$E_{UADC}=\frac{∂N}{(∂U_{ADC} )}=-m⋅\frac{Q^2}{k^2} ⋅\frac{U_{OFF}}{(U_{ADC}^2 )}-\frac{U_A}{(U_{ADC}^2)}$


$E_{UOFF}=\frac{∂N}{(∂U_{OFF} )}=m⋅\frac{Q^2}{k^2} ⋅\frac{1}{U_{ADC}} $

$E_{NA}=\frac{∂N}{(∂N_A )}=1$

Berechnung in Python


```python
"""Analytische Bechnungnung der Sensitivitäten im Arbeitspunkt Q0"""
Q0 = 0.5
EUADC = -m*Q0**2/k**2*Uoff0/UADC0**2 - NA0/UADC0
EUoff = m*Q0**2/k**2/UADC0
ENA = 1

print("EUADC: {}".format(EUADC) , "EUoff: {}".format(EUoff) , 'ENA: {} '.format(ENA))
```

c) Zur Verifikation werden Zufallszahlen erzeugt, die der definierten Verteilung entsprechen. Mit diesen Zufallszahlen wird eine statistische Simulation der Größe N durchgeführt. Mit dem Befehl polyfit kann der lineare Koeffizient des Polynoms bestimmt werden, er entspricht der oben berechneten Empfindlichkeit.

```python
"""Vergleich mit statistischer Simulation"""

M = 1000000
UADCsim= np.random.normal(UADC0, sigUADC, M)
NAsim= np.random.uniform(NA0-dNA/2, NA0+dNA/2, M)
Uoffsim = np.random.uniform(Uoff0-dUoff/2,Uoff0+dUoff/2,M)
Nsim = m*Q0**2/k**2 + m*Q0**2/k**2*Uoffsim/UADCsim + NAsim
pUADC = np.polyfit(UADCsim,Nsim,1)
EUADCsim = pUADC[0]
pUoff = np.polyfit(Uoffsim,Nsim,1)
EUoffsim = pUoff[0]
pNA = np.polyfit(NAsim,Nsim,1)
ENAsim = pNA[0]


print(f"EUADCsim: {EUADCsim}" , f"EUoffsim: {EUoffsim}" , f'ENAsim: {ENAsim} ')
```

Vergleich der Zahlenwerte bestätigt die analytische Rechnung



```python
dict={'Größe':['EUADC','EUOFF','ENA'], 'Analytische Rechnung':[EUADC,EUoff,ENA], 'Statistische Simulation':[EUADCsim,EUoffsim,ENAsim]}

pd.DataFrame(dict)
```

d) Berechnung der Toleranzen bei arithmetischer Tolerierung

$T_{ARI}=|E_{UADC}⋅ΔU_{ADC} |+|E_{UOFF}⋅ΔU_{OFF} |+|E_{NA}⋅ΔN_A |$



Um die Toleranz $Δ$Q zu berechnen wird durch die Steigung der Kennlinie dividiert.

Es ergibt sich eine Toleranz von $ΔQARI = 0.002 m³/s$.

```python
"""Arithmetische Tolerierung im Arbeitspunkt Q0"""

EN = (k**2)*UADC0/(2*m*Q0*(UADC0+Uoff0))
dNari = abs(EUADC*dUADC) + abs(ENA*dNA) + abs(EUoff*dUoff)
dQari = EN*dNari

print(f"dQari: {dQari}")
```

e) Berechnung der Standardabweichung bei statistischer Tolerierung




$σ_{STA}=\sqrt{((E_{UADC}⋅σ_{UADC} )^2+(E_{UOFF}⋅σ_{UOFF} )^2+(E_{NA}⋅σ_{NA} )^2 )}$


Um die Toleranz ΔQ zu berechnen wird wieder durch die Steigung der Kennlinie dividiert. Die Aussagesicherheit von 99.73 % entspricht der $\pm3.σ$ Toleranz, das Ergebnis muss deshalb mit 6 multipliziert werden.


```python
""" Statistische Tolerierung im Arbeitspunkt Q0"""
EN = (k**2)*UADC0/(2*m*Q0*(UADC0+Uoff0))
dNsta = ((EUADC**2)*sigUADC**2 + ((ENA)**2)*(sigNA)**2 + (EUoff**2)*sigUoff**2)**0.5
dQsta = EN*6*dNsta

print(f"dQsta: {dQsta}")
```

Es ergibt sich eine Toleranz von $ΔQ_{STA}$ = 0.0024 m³/s. Sie ist größer als die Toleranz bei arithmetischer Tolerierung.

f) Für die statistische Simulation wird der bereits berechnete Datensatz verwendet. Er wird sor-tiert und die relative kumulative Häufigkeit $F$ bestimmt. Die Toleranzen ergeben sich aus den Bedingungen 
$F_{MIN} = (1 - \gamma)/2     $          und         $F_{MAX} = (1 + \gamma)/2$.


```python
"""Prüfung der Toleranz über statistische Simulation mit berechneten Daten"""
f = (1/M)*(np.ones(np.size(Nsim)))
F = np.cumsum(f)
Nsort = np.sort(Nsim)
mmin = np.nonzero(F <(1-0.9973)/2)
mmax = np.nonzero(F >=(1+0.9973)/2)
dQsim = (Nsort[mmax[0][0]]-Nsort[mmin[0][-1]])/EN

print("dQsim: {}".format(dQsim))
```

Das Simulationsergebnis liegt mit einer Toleranz von $ΔQ_{SIM}$ = 0.0019 m³/s unterhalb von dem Wert der arithmetisch berechneten Toleranz $ΔQ_{ARI}$ = 0.002 m³/s. Zur Erklärung wird eine Darstellung der einzelnen Toleranzursachen und der Verteilungsfunktion bei der Simulation herangezogen.

```python
"""Zusammenfassung der Toleranzen als Matrix"""
dQ = [abs(EUADC*dUADC), abs(6*EUADC*sigUADC), EUoff*dUoff, 6*EUoff*sigUoff, ENA*dNA , 6*ENA*sigNA]


Fnorm = norm.cdf(Nsort,np.mean(Nsort),np.std(Nsort))

x= np.arange(3)
fig1, ax1 = plt.subplots()

plt.bar(x[0],dQ[0], width=0.25, color='b')
plt.bar(x[0]+.3,dQ[1], width=0.25, color='r')
plt.bar(x[1],dQ[2], width=0.25, color='b')
plt.bar(x[1]+0.3,dQ[3], width=0.25, color='r')
plt.bar(x[2],dQ[4], width=0.25, color='b')
plt.bar(x[2]+0.3,dQ[5], width=0.25, color='r')
plt.xticks(x+0.135, ('UADC', 'UOFF', 'NA'))
plt.ylabel('Toleranzbeitrag')
plt.xlabel('Toleranzursache')
plt.legend(('arithmetisch', 'Statistisch'))

fig2, ax2 = plt.subplots()
plt.plot(Nsort,F, '--')
plt.plot(Nsort, Fnorm, '--', color='r')
plt.ylabel('Kumulative Wahrscheinlichkeit F(N)')
plt.xlabel('ADC-Wert')
plt.legend(('Simulation', 'Normalverteilung'))
plt.show()

```

Durch die Umrechnung der Rechteckverteilung bei Offset-Spannung und Auflösung auf eine äquivalente Standardabweichung wird der Toleranzbereich mit $\gamma = 99.73$% vergrößert. Da nur wenige Maße überlagert werden, ist der Gewinn der statistischen Tolerierung gering. Damit ist die Toleranz bei statistischer Tolerierung größer als bei arithmetischer Tolerierung. Die Simulation zeigt, dass die analytisch berechnete statistische Tolerierung nicht der Realität entspricht.

g) Zur Verbesserung der Toleranz kann das Ergebnis durch die gemessene Referenzspannung geteilt werden. Allerdings muss dazu die Spannung UREF vom ADC gemessen werden. Beide Messungen haben eine voneinander unabängigen Auflösungsfehler.







 


Berechnung der Toleranzen bei arithmetischer Tolerierung



$T_{KOM}=|E_{NA1}⋅ΔNA_1 |+|E_{NA2}⋅ΔNA_2 |$


Um die Toleranz ΔQ zu berechnen wird durch die Steigung der Kennlinie dividiert. 


```python
"""Kompensation Offsetfehler Arithmetische Tolerierung im Arbeitspunkt Q0"""
ENA1 = UADC0/(UADC0+Uoff0)
ENA2 = (UADC0/(UADC0+Uoff0))**2
dNkom = abs(ENA1*dNA) + abs(ENA2*dNA)
dQkom = EN*dNkom

print("dQkom: {}".format(dQkom))
```

Es ergibt sich eine Toleranz von $ΔQ_{KOM}$ = 0.0019 m³/s. Sie ist geringfügig kleiner als die Toleranz bei arithmetischer Tolerierung ohne Rückmessung der Versorgungsspannung. Der Vorteil steigt, wenn die Offsetspannung zunimmt.
